```python
import numpy as np
import cupy as cp
from scipy.signal import convolve2d
import scipy.ndimage
import matplotlib.pyplot as plt
import time
from IPython import display
```

# Aberration correction simulation

```python
xp = np
```

Set up arrays

```python
N = 256
x, y = xp.meshgrid(xp.linspace(-1, 1, N), xp.linspace(-1, 1, N))
r = xp.sqrt(x ** 2 + y ** 2)
theta = xp.arctan2(y, x)
circ = r < 1
g = xp.exp(-r ** 2 / 2 / (0.5 ** 2))
G = xp.zeros((int(2 * N), int(2 * N)))
G[int(N / 2): int(3 * N / 2), int(N / 2): int(3 * N / 2)] = g
normval = xp.sum(circ)
```

Zernike aberrations

```python
Z00 = xp.ones_like(x) * circ
Z11 = 2 * r * xp.cos(theta) * circ
Z1_1 = 2 * r * xp.sin(theta) *circ
Z20 = xp.sqrt(3) * (2 * r ** 2 - 1) * circ
# Z22 = sqrt(6)*r.^2.*cos(2*theta).*circ
# Z2_2 = sqrt(6)*r.^2.*sin(2*theta).*circ
# Z31 = 2*sqrt(2)*(3*r.^3-2*r).*cos(theta).*circ
# Z3_1 = 2*sqrt(2)*(3*r.^3-2*r).*sin(theta).*circ
# Z33 = 2*sqrt(2)*r.^3.*cos(3*theta).*circ
# Z3_3 = 2*sqrt(2)*r.^3.*sin(3*theta).*circ
# Z40 = sqrt(5)*(6*r.^4-6*r.^2+1).*circ
# Z42 = sqrt(10)*(4*r.^4-3*r.^2).*cos(2*theta).*circ
# Z4_2 = sqrt(10)*(4*r.^4-3*r.^2).*sin(2*theta).*circ
# Z44 = sqrt(10)*r.^4.*cos(4*theta).*circ
# Z4_4 = sqrt(10)*r.^4.*sin(4*theta).*circ
# Z51 = 2*sqrt(3)*(10*r.^5-12*r.^3+3*r).*cos(theta).*circ
# Z5_1 = 2*sqrt(3)*(10*r.^5-12*r.^3+3*r).*sin(theta).*circ
# Z53 = 2*sqrt(3)*(5*r.^5-4*r.^3).*cos(3*theta).*circ
# Z5_3 = 2*sqrt(3)*(5*r.^5-4*r.^3).*sin(3*theta).*circ
# Z55 = sqrt(12)*r.^5.*cos(5*theta).*circ
# Z5_5 = sqrt(12)*r.^5.*sin(5*theta).*circ
```

## Simulated PSF images (input)

Create a random image with low spatial frequency content for aberration to be measured

```python
ri0 = xp.random.rand(N, N) - 0.5
kern = xp.reshape(xp.exp(-r ** 2 / (2 * 0.35 ** 2)), (N, N))
if xp == cp:
    ri = xp.array(convolve2d(ri0.get(), kern.get(), mode='same')) * circ / 1
else:
    ri = convolve2d(ri0, kern, mode='same') * circ / 1
# % ri = imresize(ri,10,'bicubic')
# % ri = 1.0*ri(9*N/2:11*N/2-1,9*N/2:11*N/2-1).*circ+atan2(y-0.5,x-0.3).*circ-atan2(y+0.4,x-0.3).*circ
# % ri = (2*ri(9*N/2:11*N/2-1,9*N/2:11*N/2-1)+2*Z40+2*Z11+1*Z2_2+2*Z11).*circ
# % ri = (30 * Z22 + 2 * Z2_2 - 5 * Z31 + 7 * Z3_1 + 5 * Z33 - 2 * Z3_3).*circ/2

# for the starting test aberration
x2, y2 = xp.meshgrid(xp.linspace(-0.2, 0.2, N), xp.linspace(-0.2, 0.2, N))
r2 = xp.sqrt(x2 ** 2 + y2 ** 2)
ri2 = xp.random.rand(N, N) - 0.5
kern = xp.exp(-r2 ** 2 / (2 * 0.35 ** 2))
if xp == cp:
    ri2 = xp.array(convolve2d(ri2.get(), kern.get()))
else:
    ri2 = convolve2d(ri2, kern)
# ri2 = xp.reshape(ri2,ri2.shape * 10)
ri2 = 1.0 * ri2[int(N / 2):int(3 * N / 2), int(N / 2):int(3 * N / 2)] * circ
```

Work out measured PSFs for the 3 applied masks

```python
plt.figure(100)
if xp == cp:
    plt.imshow((xp.angle(xp.exp(1j * ri))).get())
else:
    plt.imshow(xp.angle(xp.exp(1j * ri)))
plt.title('Random phase')

sumc = xp.sum(circ)
piston = xp.sum(xp.sum(ri * circ)) / sumc
diff = (ri-piston) * circ
rms = xp.sqrt(xp.sum(diff ** 2) / sumc)
# imshow(diff.*circ,[])
# text(0,-10,sprintf('max-min=%f, rms=%f, rmswaves=%f',max(diff(:))-min(diff(:)),rms,rms/2/pi))

plt.figure(101)
hlx = theta * circ  # helical wavefront
if xp == cp:
    plt.imshow(hlx.get())
else:
    plt.imshow(hlx)
plt.title('Phase singularity')

plt.figure(102)
defocus = 10 * Z20
if xp == cp:
    plt.imshow(defocus.get())
else:
    plt.imshow(defocus)
plt.title('Defocus')

#% ampl=circ
#% ampl=circ.*(1+x/3);  % a tilt in intensity accross the pupil
ampl = circ * xp.exp(-((x - 0.25) ** 2 + 2 * y ** 2) / 2 / 1 ** 2)  # funny gaussian illumination to simulater a laser beam scenario; sigma=1
```

```python
nphot = 1e8  # Number of photons per detected image
bg = xp.ones((int(2*N), int(2*N)))  # background
```

Apply three masks

```python
masks = [ri, ri + defocus, ri-defocus]
titles = ['PSF', 'PSF+defocus', 'PSF-defocus']
phase = [None, None, None]
for i in range(3):
    p1 = xp.fft.fftshift(xp.fft.fft2(ampl * xp.exp(1j * masks[i]), (2*N, 2*N)))
    p1i = abs(p1) ** 2 / xp.sum(abs(p1) ** 2) * nphot + bg
    p1i_n = (xp.random.poisson(p1i) - bg) * G
    ps = p1i_n[int(N/2-1):int(3*N/2-1), int(N/2-1):int(3*N/2-1)]
    plt.figure(111 + i)
    plt.title(titles[i])
    if xp == cp:
        plt.imshow(ps.get())
    else:
        plt.imshow(ps)
    phase[i] = p1i_n + 0j
```

```python
cumphase = xp.zeros((N, N))
Q = circ * xp.exp(1j * ri2)
```

## Iteration loop

```python
def show_animation(i, img):
    ax = plt.subplot(2,6,i)
    ax.clear()
    if xp == cp:
        ax.imshow(img.get())
    else:
        ax.imshow(img)
    ax.set_title(f"frame {i}")
    display.clear_output(wait=True)
    display.display(ax.get_figure())
```

```python
t0 = time.time()
nits = 1
n = 40
while nits < 500:
    q1 = xp.fft.fftshift(xp.fft.fft2(Q, (2*N, 2*N)))
    q2 = xp.fft.fftshift(xp.fft.fft2(Q * xp.exp(1j * defocus), (2*N, 2*N)))
    q3 = xp.fft.fftshift(xp.fft.fft2(Q * xp.exp(1j * (-defocus)), (2*N, 2*N)))
    q1s = q1[int(1*N/2-1):int(3*N/2), int(1*N/2-1):int(3*N/2)]
    if xp.mod(nits, n) == 0:
        show_animation(1,abs(q1s) ** 2)

    Q1 = xp.fft.ifft2(xp.fft.fftshift(xp.sqrt(phase[0]) * xp.exp(1j * xp.angle(q1))))
    Q1[:N, :N] = Q1[:N, :N] * circ
    Q2 = xp.fft.ifft2(xp.fft.fftshift(xp.sqrt(phase[1]) * xp.exp(1j * xp.angle(q2))))
    Q2[:N, :N] = Q2[:N, :N] * circ * xp.exp(-1j * defocus)
    Q3 = xp.fft.ifft2(xp.fft.fftshift(xp.sqrt(phase[2]) * xp.exp(1j * xp.angle(q3))))
    Q3[:N, :N] = Q3[:N, :N] * circ * xp.exp(1j * defocus)

    Q = Q1[:N, :N] + Q2[:N, :N] + Q3[:N, :N]

# %     Q=Q2(1:N,1:N)+Q3(1:N,1:N)

    if xp.mod(nits, n) == 0:
        if xp.mod(nits, n) == 0:
            show_animation(2,xp.angle(Q))

    sumc = xp.sum(circ)
    diff = xp.angle(Q * xp.exp(-1j * ri))
    piston = xp.sum(diff * circ) / sumc
    diff = (diff-piston) * circ
    rms= xp.sqrt(xp.sum(diff ** 2) / sumc)

    if xp.mod(nits,n) == 0:
        show_animation(3,diff * circ)

    if xp.mod(nits,n) == 0:
        show_animation(4,xp.sqrt(abs(Q)))
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

#         plt.subplot(2,6,5)
#         plt.plot(xp.squeeze(abs(Q[N//2,:])))

    # Constrain intensity in pupil to have slowly varying values by fitting to Zernike polynomials, up to radial order
    # 4 to help with gaussian illumination
    Qa = abs(Q)
    Qb = sum(Qa * Z00) * Z00
    Qb = Qb + sum(Qa * Z11) * Z11
    Qb = Qb + (Qa * Z1_1) * Z1_1
    Qb = Qb + sum(Qa * Z20) * Z20
#
#     Qb=Qb+sum(sum(Qa.*Z22))*Z22
#
#     Qb=Qb+sum(sum(Qa.*Z2_2))*Z2_2
#
#     Qb=Qb+sum(sum(Qa.*Z31))*Z31;
#
#     Qb=Qb+sum(sum(Qa.*Z3_1))*Z3_1;
#
#     Qb=Qb+sum(sum(Qa.*Z40))*Z40;
#
#     Qb=Qb+sum(sum(Qa.*Z42))*Z42;
#
#     Qb=Qb+sum(sum(Qa.*Z4_2))*Z4_2;
#
#     Qb=Qb+sum(sum(Qa.*Z44))*Z44;
#
#     Qb=Qb+sum(sum(Qa.*Z4_4))*Z4_4;

    Qc = xp.angle(Q)
    Q = Qb * xp.exp(1j*Qc) / sumc
    if xp.mod(nits, n) == 0:
        q = xp.angle(Q)
        q[N//2:, :] = xp.unwrap(q[N//2:, :], axis=0)
        q = xp.flipud(q)
        q[N//2:, :] = xp.unwrap(q[N//2:, :], axis=0)
        q[:, N//2:] = xp.unwrap(q[:, N//2:], axis=1)
        q = xp.fliplr(q)
        q[:, N//2:] = xp.unwrap(q[:, N//2:], axis=1)
        q = xp.flipud(xp.fliplr(q))
        show_animation(6,q * circ)
        plt.text(0, -10, 'nits=%d , rms=%f, rmswaves=%f' % (nits, rms, rms / 2 / xp.pi))


        # Plot poles in phase function
        # plt.figure(211)
        Qc2 = xp.exp(1j * Qc)
        fr = xp.exp(1j * xp.array([(xp.pi/4, 0, -xp.pi/4), (xp.pi/2, 0, -xp.pi/2), (3*xp.pi/4, xp.pi, -3*xp.pi/4)]))
        fr[1, 1] = 0
        fl = xp.conj(fr)
        # Qcpr = abs(imfilter(Qc2,fr))
        # Qcpl=abs(imfilter(Qc2,fl))       ??????
        Qcpr = abs(scipy.ndimage.correlate(Qc2, fr, mode='constant').transpose())
        Qcpl = abs(scipy.ndimage.correlate(Qc2, fl, mode='constant').transpose())
        Qcr = xp.reshape([Qcpl, Qcpr, xp.zeros((N, N))], (3, N, N))
        Qcr = Qcr / Qcr.max()
        # plt.imshow(Qcr)

# % knock out singularities with conjugates
# %         Qr=Qcpr.^3
# %         Qr(r>0.9)=0
# %         pr=reshape(FastPeakFind(Qr,30),2,[]).'
# %         for j = 1:size(pr,1)
#
# %             Q=Q.*exp(1i*atan2(x-(pr(j,1)-128)/128,y-(pr(j,2)-128)/128));
#
# %         end
#
# %         Ql=Qcpl.^3;
#
# %         Ql(r>0.95)=0;
#
# %         pl=reshape(FastPeakFind(Ql,30),2,[]).';
#
# %         for j = 1:size(pl,1)
#
# %             Q=Q.*exp(-1i*atan2(x-(pl(j,1)-128)/128,y-(pl(j,2)-128)/128))
# %         end
#
#         % unwrap phase
# %         q=angle(Q);

        q[N // 2:, :] = xp.unwrap(q[N // 2:, :], axis=0)
        q = xp.flipud(q)
        q[N // 2:, :] = xp.unwrap(q[N // 2:, :], axis=0)
        q[:, N // 2:] = xp.unwrap(q[:, N // 2:], axis=1)
        q = xp.fliplr(q)
        q[:, N // 2:] = xp.unwrap(q[:, N // 2:], axis=1)
        q = xp.flipud(xp.fliplr(q))

        diff = q
        piston = xp.sum(diff *circ)/sumc
        diff = (diff-piston) * circ
        rms = xp.sqrt(xp.sum(diff ** 2) / sumc)
        show_animation(7,q * circ)
        plt.text(0, -10, 'nits=%d , rms=%f, rmswaves=%f' % (nits, rms, rms / 2 / xp.pi))
        Qc = q

        # Constrain phase in pupil to have slowly varying values by fitting to Zernike
        Qd = xp.sum(Qc * Z00)*Z00
        Qd = Qd+ xp.sum(Qc * Z11)*Z11
        Qd = Qd+ xp.sum(Qc * Z1_1)*Z1_1
        Qd = Qd+ xp.sum(Qc * Z20)*Z20
        # Qd=Qd+sum(Qc.*Z22, 'all')*Z22
        # Qd=Qd+sum(Qc.*Z2_2, 'all')*Z2_2;
#
#         Qd=Qd+sum(Qc.*Z31, 'all')*Z31;
#
#         Qd=Qd+sum(Qc.*Z3_1, 'all')*Z3_1;
#
#         Qd=Qd+sum(Qc.*Z33, 'all')*Z33;
#
#         Qd=Qd+sum(Qc.*Z3_3, 'all')*Z3_3;
#
#         Qd=Qd+sum(Qc.*Z40, 'all')*Z40;
#
#         Qd=Qd+sum(Qc.*Z42, 'all')*Z42;
#
#         Qd=Qd+sum(Qc.*Z4_2, 'all')*Z4_2;
#
#         Qd=Qd+sum(Qc.*Z44, 'all')*Z44;
#
#         Qd=Qd+sum(Qc.*Z4_4, 'all')*Z4_4;
#
#         Qd=Qd+sum(Qc.*Z51, 'all')*Z51;
#
#         Qd=Qd+sum(Qc.*Z5_1, 'all')*Z5_1;
#
#         Qd=Qd+sum(Qc.*Z53, 'all')*Z53;
#
#         Qd=Qd+sum(Qc.*Z5_3, 'all')*Z5_3;
#
#         Qd=Qd+sum(Qc.*Z55, 'all')*Z55;
#
#         Qd=Qd+sum(Qc.*Z5_5, 'all')*Z5_5;
        Qd=Qd/normval

        fac = 2
        Q = abs(Q) * xp.exp(1j * Qd * (1-1 / fac))
        # Q = abs(Q).*exp(1i*Qd)
        show_animation(8,Qd)
# %         def = 0.9 * def;

        cumphase = cumphase + Qd/fac
        show_animation(9,xp.angle(xp.exp(1j * cumphase)))

#         new_masks = [ri - cumphase, ri - cumphase + defocus, ri - cumphase - defocus]
#         for i in range(3):
#             p1 = xp.fft.fftshift(xp.fft.fft2(ampl * xp.exp(1j * new_masks[i]), (2 * N, 2 * N)))
#             p1i = abs(p1) ** 2 / xp.sum(abs(p1) ** 2) * nphot + bg
#             p1i_n = (xp.random.poisson(p1i) - bg) * G
#             ps = p1i_n[int(N / 2 - 1):int(3 * N / 2 - 1), int(N / 2 - 1):int(3 * N / 2 - 1)]
#             show_animation(9+i,ps)
#             plt.title(titles[i])
#             phase[i] = p1i_n + 0j

    nits = nits + 1

print(f'Elapsed time: {time.time() - t0}s')
```
